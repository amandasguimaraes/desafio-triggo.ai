O conjunto de dados de vendas do Olist é uma coleção de dados anonimizados sobre pedidos feitos no Olist entre setembro de 2016 e setembro de 2018.

Este projeto foi desenvolvido utilizando a base de dados "Brazilian E-commerce Public Dataset by Olist" disponibilizada no Kaggle.

### Limpeza e Verificação dos Dados

```python
# Verificando valores nulos em cada dataset
for name, df in zip(
    ['customers', 'geolocation', 'order_items', 'order_payments', 'order_reviews', 'orders', 'products', 'sellers'],
    [customers, geolocation, order_items, order_payments, order_reviews, orders, products, sellers]
):
    print(f'{name}:')
    print(df.isnull().sum())
    print()
```

```python
# Convertendo colundas de datas para o formato datatime
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])

# Removendo produtos sem categoria
products = products.dropna(subset=['product_category_name'])

# Removendo duplicatas
for df in [customers, geolocation, order_items, order_payments, order_reviews, orders, products, sellers]:
    df.drop_duplicates(inplace=True)

# Adicionando tradução das categorias de produto
products = products.merge(category_translation, on='product_category_name', how='left')
```

### Criação do Modelo Relacional

```python
# Criando uma base unificada para análise
df = orders.merge(order_items, on='order_id') \
           .merge(order_payments, on='order_id') \
           .merge(order_reviews, on='order_id', how='left') \
           .merge(customers, on='customer_id') \
           .merge(sellers, on='seller_id') \
           .merge(products, on='product_id')

df['order_purchase_month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)

df.info()
df.head()
```

#### a) Volume de pedidos por mês: existe sazonalidade nas vendas?

```python
query = """
SELECT order_purchase_month, COUNT(DISTINCT order_id) as total_orders
FROM olist_data
GROUP BY order_purchase_month
ORDER BY order_purchase_month;
"""

orders_per_month = pd.read_sql_query(query, conn)

plt.figure(figsize=(15, 6))
sns.lineplot(data=orders_per_month, x='order_purchase_month', y='total_orders', marker='o')
plt.title('Volume de Pedidos por Mês')
plt.xticks(rotation=45)
plt.xlabel('Mês')
plt.ylabel('Número de Pedidos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### b) Distribuição do tempo de entrega dos pedidos

```python
df['delivery_time_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

plt.figure(figsize=(12, 6))
sns.histplot(df['delivery_time_days'].dropna(), bins=30, kde=True)
plt.title('Distribuição do Tempo de Entrega (dias)')
plt.xlabel('Dias para Entrega')
plt.ylabel('Quantidade de Pedidos')
plt.grid(True)
plt.show()
```

#### c) Relação entre o valor do frete e a distância de entrega

```python
# Calculando distância aproximada entre cliente e vendedor
from geopy.distance import geodesic

# Agrupando coordenadas médias por prefixo de CEP (para clientes e vendedores)
geo_customers = geolocation.groupby('geolocation_zip_code_prefix')[['geolocation_lat', 'geolocation_lng']].mean()
geo_sellers = geolocation.groupby('geolocation_zip_code_prefix')[['geolocation_lat', 'geolocation_lng']].mean()

# Juntando com as coordenadas
df_geo = df.copy()
df_geo = df_geo.merge(geo_customers, how='left', left_on='customer_zip_code_prefix', right_index=True)
df_geo = df_geo.merge(geo_sellers, how='left', left_on='seller_zip_code_prefix', right_index=True, suffixes=('_cust', '_sell'))

# Calculando distância
def calc_distance(row):
    try:
        return geodesic((row['geolocation_lat_cust'], row['geolocation_lng_cust']),
                        (row['geolocation_lat_sell'], row['geolocation_lng_sell'])).kilometers
    except:
        return np.nan

df_geo['distance_km'] = df_geo.apply(calc_distance, axis=1)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_geo.sample(5000), x='distance_km', y='freight_value', alpha=0.4)
plt.title('Valor do Frete vs Distância de Entrega')
plt.xlabel('Distância (km)')
plt.ylabel('Valor do Frete (R$)')
plt.grid(True)
plt.show()
```

#### Análise de Retenção de Clientes

```python
# Carregar os dados
orders = pd.read_csv('data/olist_orders_dataset.csv')
customers = pd.read_csv('data/olist_customers_dataset.csv')

# Combinar os dados de pedidos e clientes
orders_customers = pd.merge(orders, customers, on='customer_id')

# Contar o número de pedidos por cliente
customer_order_counts = orders_customers.groupby('customer_unique_id')['order_id'].nunique()

# Identificar clientes recorrentes
recurring_customers = customer_order_counts[customer_order_counts > 1]

# Calcular a taxa de retenção
retention_rate = len(recurring_customers) / customer_order_counts.count()

print(f"Taxa de clientes recorrentes: {retention_rate:.2%}")
```

```python
**Insight:** Uma taxa de retenção baixa pode indicar a necessidade de estratégias para fidelizar clientes, como programas de fidelidade ou promoções exclusivas para clientes recorrentes.
```

#### Segmentação de Clientes (Clustering)

```python
# Carregar os dados
orders = pd.read_csv('data/olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])
order_items = pd.read_csv('data/olist_order_items_dataset.csv')
customers = pd.read_csv('data/olist_customers_dataset.csv')

# Combinar os dados
orders_items = pd.merge(orders, order_items, on='order_id')
orders_items_customers = pd.merge(orders_items, customers, on='customer_id')

# Calcular métricas por cliente
customer_metrics = orders_items_customers.groupby('customer_unique_id').agg({
    'order_id': 'nunique',
    'price': 'sum',
    'order_purchase_timestamp': ['min', 'max']
})

customer_metrics.columns = ['order_count', 'total_spent', 'first_purchase', 'last_purchase']
customer_metrics['customer_lifetime'] = (customer_metrics['last_purchase'] - customer_metrics['first_purchase']).dt.days

# Preencher valores nulos
customer_metrics['customer_lifetime'] = customer_metrics['customer_lifetime'].fillna(0)

# Selecionar características para clustering
features = customer_metrics[['order_count', 'total_spent', 'customer_lifetime']]

# Aplicar K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
customer_metrics['cluster'] = kmeans.fit_predict(features)

# Visualizar os clusters
plt.scatter(customer_metrics['order_count'], customer_metrics['total_spent'], c=customer_metrics['cluster'])
plt.xlabel('Número de Pedidos')
plt.ylabel('Total Gasto')
plt.title('Segmentação de Clientes')
plt.show()
```

#### Análise de Satisfação do Cliente

```python
# Carregar os dados
reviews = pd.read_csv('data/olist_order_reviews_dataset.csv')
orders = pd.read_csv('data/olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp', 'order_delivered_customer_date'])
order_items = pd.read_csv('data/olist_order_items_dataset.csv')
products = pd.read_csv('data/olist_products_dataset.csv')

# Combinar os dados
orders_reviews = pd.merge(orders, reviews, on='order_id')
orders_reviews_items = pd.merge(orders_reviews, order_items, on='order_id')
orders_reviews_items_products = pd.merge(orders_reviews_items, products, on='product_id')

# Calcular tempo de entrega
orders_reviews_items_products['delivery_time'] = (orders_reviews_items_products['order_delivered_customer_date'] - orders_reviews_items_products['order_purchase_timestamp']).dt.days

# Analisar a relação entre avaliação e tempo de entrega
sns.boxplot(x='review_score', y='delivery_time', data=orders_reviews_items_products)
plt.title('Tempo de Entrega por Avaliação')
plt.xlabel('Avaliação')
plt.ylabel('Tempo de Entrega (dias)')
plt.show()

# Analisar a média de avaliação por categoria de produto
avg_review_by_category = orders_reviews_items_products.groupby('product_category_name')['review_score'].mean().sort_values(ascending=False)

# Visualizar as categorias com melhor avaliação
avg_review_by_category.head(10).plot(kind='barh')
plt.title('Média de Avaliação por Categoria de Produto')
plt.xlabel('Média de Avaliação')
plt.ylabel('Categoria de Produto')
plt.show()
```

```python
**Insight:** Identificar categorias de produtos com avaliações consistentemente baixas pode indicar problemas de qualidade ou expectativas não atendidas. Além disso, tempos de entrega mais longos podem estar correlacionados com avaliações mais baixas, destacando a importância da logística eficiente.
```
