```python
# Importando bibliotecas
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo constante de caminho
INTERMEDIATE_PATH = '../data/intermediate/'

# Verificando arquivos dispon√≠veis
try:
    arquivos = [f for f in os.listdir(INTERMEDIATE_PATH) if f.endswith('.csv')]
    print(f"üìÅ {len(arquivos)} arquivos encontrados para an√°lise.")
except FileNotFoundError:
    print(f"‚ùå Diret√≥rio {INTERMEDIATE_PATH} n√£o encontrado. Verifique o caminho.")
    arquivos = []
```

```python
# Definindo fun√ß√£o explorat√≥ria de dados
def explorar_dataframe(df, nome):
    print(f'\n{"="*50}')
    print(f'üìò Explorando: {nome.upper()}')
    print(f'{"="*50}')
    
    # Informa√ß√µes b√°sicas
    print(f'\nüîç Dimens√µes: {df.shape[0]} linhas x {df.shape[1]} colunas')
    
    # Amostra dos dados
    print('\nüìã Primeiras linhas:')
    display(df.head())
    
    # Informa√ß√µes sobre tipos de dados
    print('\n‚ÑπÔ∏è Informa√ß√µes sobre tipos de dados:')
    df.info()
    
    # Resumo estat√≠stico
    print('\nüìä Estat√≠sticas descritivas:')
    display(df.describe(include='all').T)
```

```python
# Carregando e explorando todas as tabelas
if arquivos:
    for arquivo in arquivos:
        nome_df = arquivo.replace('.csv', '')
        try:
            df = pd.read_csv(os.path.join(INTERMEDIATE_PATH, arquivo))
            explorar_dataframe(df, nome_df)
        except Exception as e:
            print(f"‚ùå Erro ao processar {arquivo}: {e}")
```
